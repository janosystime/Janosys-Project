## Importação das bibliotecas Pandas e Plotly.

```python
import pandas as pd
import plotly.express as px
```

##Importação dos dados do GitHub

```python
url = 'https://github.com/janosystime/Janosys-Project/blob/main/bd/source/tabelaAPIMockada.csv?raw=true'

df = pd.read_csv(url, encoding="latin1", sep="\t")
```

##Exibição do DataFrame Bruto.

```python
display(df.tail())
```

##Filtragem e exibição do Dataframe por Região e Faixa-Etária

```python
df_regiao_faixas = df[['Região','Criança 2010','Adolescente 2010','Adulto 2010','Idoso 2010',
                                'Criança 2022', 'Adolescente 2022','Adulto 2022','Idoso 2022']]
display(df_regiao_faixas)
```

##Enumerando regiões e feixa-etárias disponíveis, e exbindo para o usuário.

```python
print("Regiões disponíveis:")
for i, regiao in enumerate(df["Região"], start=1):
    print(f"{i}: {regiao}")

colunas = df_regiao_faixas.columns.tolist()
faixas = colunas[1:]
print("\nFaixas etárias disponíveis:")
for i, faixa in enumerate(faixas, start=1):
    print(f"{i}: {faixa}")
```

##Requisição de filtro desejado pelo usuário.

```python
regiao_idx = int(input("\nEscolha o número da região: ")) - 1
faixa_idx = int(input("Escolha o número da faixa etária: ")) - 1

regiao_selecionada = df["Região"].iloc[regiao_idx]
faixa_selecionada = faixas[faixa_idx]
```

##Pegando os dados com base no filtro escolhido pelo usuários.

```python
populacao = df.loc[df["Região"] == regiao_selecionada, faixa_selecionada].values[0]

print(f"\nA população de {faixa_selecionada} na região {regiao_selecionada} é: {populacao}")
```

##Definindo gráfico no plotly com base nos filtros escolhidos pelo usuário.

```python
fig = px.bar(
    x=[regiao_selecionada],
    y=[populacao],
    labels={'x':'Região', 'y':'População'},
    title=f'População de {faixa_selecionada} na região {regiao_selecionada}'
)
```

##Exibição do gráfico gerado.

```python

fig.show()
```
